```julia
using Plots   
pyplot() 
using LaTeXStrings  
using HDF5 
```

```julia
#read the ldos data
ldos_tot_re = h5open("billion_250cheby_domain_U_5.h5", "r") do file
 
    ldos_tot_re = read(file, "ldos");
 
end
```

```julia
num_frag = 70
xvals = -num_frag/2:num_frag/2 - 1
yvals = range(-0.99, 0.99; length=250)

quantity = Int(floor(log10(100000000)))  # Convert log value to integer

# Identify indices for xticks near the edges and 0
edge_offset = Int(round(length(xvals) * 0.15))  
index_left = Int(round(edge_offset))
index_right = Int(round(length(xvals) - edge_offset))
index_center = Int(round(length(xvals) / 2))

# Map indices to values
selected_xvals = [xvals[index_left], 0, xvals[index_right]]

default(dpi=300) 

# Directly define LaTeX-formatted labels for the xticks
formatted_labels = [
    L"-5 \times 10^{%$quantity}", 
    L"0", 
    L"5 \times 10^{%$quantity}"
]
vmin, vmax = extrema(ldos_tot_re)  

# Set high DPI for better quality
#

# Create contour plot
p = contourf(
    xvals,
    yvals .* 10,
    ldos_tot_re,
    ylabel="Energy [t]",
    levels=500,
    c=:inferno,
 
    linewidth=0.0,
    xticks=(selected_xvals, formatted_labels),  # Now LaTeX labels should work
    tick_direction=:out,
    ylims=(-4.25, 4.25),
    clim=(vmin, vmax),  
    tickfontsize=27,  
    guidefontsize=27,
    colorbar_ticks=[],  # Hide colorbar tick labels
)

 
```

```julia
savefig(p, "billion_250cheby_5_U_domain.png")  
```
